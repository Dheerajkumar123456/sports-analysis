```python
def convert(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1]
        
        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
fifa['Value']=fifa['Value'].apply(convert)
fifa['Wage'] = fifa['Wage'].apply(convert)
fifa
```

```python
fifa.head(4380)
```

```python
def date_conversion(a):

    if(pd.isna(a))!= True:
        val = []
        val = a.split(',')
        dat = int(val[1])
        return dat
    else:
#         return 0
        return a

fifa['Joined']=fifa['Joined'].apply(date_conversion)
```

```python
fifa['Contract Valid Until'] = pd.to_datetime(fifa['Contract Valid Until'])
```

```python
fifa['Contract Valid Until']=fifa['Contract Valid Until'].dt.year
```

```python
fifa
```

```python
def height_conversion(height):
    if(pd.isna(height))!= True:
        h = []
        h = height.split("'")  
        ft = float(h[0])
        if( h[1] != ''):
            inch = float(h[1])
        else:
            inch = 0
        tot_inc = inch + ft*12
        h = tot_inc * 2.54
        return h
    else:
        return height
    
fifa['Height'] = fifa['Height'].apply(height_conversion)
```

```python
fifa
```

```python
fifa['Weight']=fifa['Weight'].str.replace('lbs','')
```

```python
def weight_conversion(weight):
    if(pd.isna(weight))!= True:
        w = int(weight)*0.453592
        return w
    else:
        return weight
    
fifa['Weight'] = fifa['Weight'].apply(weight_conversion)
```

```python
fifa
```

```python
def fun_convert(df_value):
    
    if(pd.isna(df_value))!= True:
        try:
            value = float(df_value[1:-1])
            suffix = df_value[-1]

            if suffix == 'M':
                value = value * 1000000
            elif suffix == 'K':
                value = value * 1000
        except ValueError:
            value = 0
        return value
   
```

```python
fifa['Release Clause']=fifa['Release Clause'].apply(fun_convert)
```

```python
fifa
```

```python
fifa['Weight'] = fifa['Weight'].fillna(value = fifa['Weight'].mean())
```

```python
fifa['Height'] = fifa['Height'].fillna(value = fifa['Height'].mean())
```
